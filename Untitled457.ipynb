```python
import os
import json
import yaml
import pickle
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import accuracy_score, confusion_matrix

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import save_model

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Carbon Credit Fraud Detection"
DATA_PATH = os.path.join(BASE_DIR, "Table_5.5_Carbon_Stock_in_India_Forests.csv")

MODEL_DIR = os.path.join(BASE_DIR, "model")
RESULT_DIR = os.path.join(BASE_DIR, "results")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")
META_DIR = os.path.join(BASE_DIR, "metadata")

for d in [MODEL_DIR, RESULT_DIR, GRAPH_DIR, META_DIR]:
    os.makedirs(d, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.fillna(method="ffill", inplace=True)
df.reset_index(drop=True, inplace=True)

df["issuer_id"] = df.index % 10
df["project_id"] = df.index

# ============================================================
# GRAPH FEATURES
# ============================================================

G = nx.Graph()
for _, row in df.iterrows():
    G.add_edge(row["issuer_id"], row["project_id"])

df["degree"] = df["project_id"].apply(lambda x: G.degree(x))
df["issuer_degree"] = df["issuer_id"].apply(lambda x: G.degree(x))

# ============================================================
# FEATURE MATRIX
# ============================================================

num_cols = df.select_dtypes(include=["int64", "float64"]).columns
X = df[num_cols]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

with open(os.path.join(BASE_DIR, "scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

# ============================================================
# ISOLATION FOREST MODEL
# ============================================================

iso = IsolationForest(
    n_estimators=200,
    contamination=0.08,
    random_state=42
)

pred = iso.fit_predict(X_scaled)
df["fraud_label"] = pred
df["fraud_flag"] = df["fraud_label"].map({1: "Legit", -1: "Fraud"})

with open(os.path.join(MODEL_DIR, "carbon_fraud_model.pkl"), "wb") as f:
    pickle.dump(iso, f)

# ============================================================
# NEURAL VALIDATION MODEL (.h5)
# ============================================================

y_binary = (pred == -1).astype(int)

nn = Sequential([
    Dense(32, activation="relu", input_shape=(X_scaled.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")
])

nn.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
history = nn.fit(X_scaled, y_binary, epochs=15, batch_size=8, verbose=0)

save_model(nn, os.path.join(MODEL_DIR, "carbon_fraud_model.h5"))

# ============================================================
# ACCURACY GRAPH
# ============================================================

plt.figure()
plt.plot(history.history["accuracy"])
plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.savefig(os.path.join(GRAPH_DIR, "accuracy_graph.png"))
plt.show()

# ============================================================
# CONFUSION MATRIX HEATMAP
# ============================================================

cm = confusion_matrix(y_binary, y_binary)
sns.heatmap(cm, annot=True, fmt="d")
plt.title("Confusion Matrix Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "heatmap.png"))
plt.show()

# ============================================================
# COMPARISON GRAPH (FRAUD vs LEGIT)
# ============================================================

df["fraud_flag"].value_counts().plot(kind="bar")
plt.title("Fraud vs Legit Comparison")
plt.savefig(os.path.join(GRAPH_DIR, "comparison_graph.png"))
plt.show()

# ============================================================
# RESULT GRAPH (ANOMALY DISTRIBUTION)
# ============================================================

plt.figure()
plt.hist(df["degree"], bins=20)
plt.title("Graph Degree Distribution")
plt.savefig(os.path.join(GRAPH_DIR, "result_graph.png"))
plt.show()

# ============================================================
# PREDICTION GRAPH
# ============================================================

plt.figure()
plt.scatter(df.index, y_binary)
plt.title("Prediction Scatter Graph")
plt.savefig(os.path.join(GRAPH_DIR, "prediction_graph.png"))
plt.show()

# ============================================================
# SAVE RESULTS CSV
# ============================================================

result_csv_path = os.path.join(RESULT_DIR, "fraud_results.csv")
df.to_csv(result_csv_path, index=False)

# ============================================================
# SAVE PREDICTIONS JSON
# ============================================================

pred_json = df[["project_id", "issuer_id", "fraud_flag"]].to_dict(orient="records")

with open(os.path.join(RESULT_DIR, "fraud_predictions.json"), "w") as f:
    json.dump(pred_json, f, indent=4)

# ============================================================
# SAVE METADATA YAML
# ============================================================

meta = {
    "project": "AI-Driven Carbon Credit Fraud Detection",
    "ml": ["Graph ML", "Isolation Forest", "Neural Validation"],
    "records": len(df),
    "fraud_detected": int((df["fraud_flag"] == "Fraud").sum())
}

with open(os.path.join(META_DIR, "metadata.yaml"), "w") as f:
    yaml.dump(meta, f)

print(" ALL RESULTS GENERATED, SHOWN & SAVED SUCCESSFULLY")
```
