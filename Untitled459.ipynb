```python
import os
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import save_model

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Carbon Credit Fraud Detection"
DATA_PATH = os.path.join(BASE_DIR, "Table_5.5_Carbon_Stock_in_India_Forests.csv")

MODEL_DIR = os.path.join(BASE_DIR, "model")
RESULT_DIR = os.path.join(BASE_DIR, "results")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")
META_DIR = os.path.join(BASE_DIR, "metadata")

for d in [MODEL_DIR, RESULT_DIR, GRAPH_DIR, META_DIR]:
    os.makedirs(d, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.fillna(method="ffill", inplace=True)
df.reset_index(drop=True, inplace=True)

df["issuer_id"] = df.index % 10
df["project_id"] = df.index

# ============================================================
# GRAPH FEATURES
# ============================================================

G = nx.Graph()
for _, r in df.iterrows():
    G.add_edge(r["issuer_id"], r["project_id"])

df["degree"] = df["project_id"].apply(lambda x: G.degree(x))
df["issuer_degree"] = df["issuer_id"].apply(lambda x: G.degree(x))

# ============================================================
# FEATURE MATRIX
# ============================================================

num_cols = df.select_dtypes(include=["int64", "float64"]).columns
X = df[num_cols]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

with open(os.path.join(BASE_DIR, "cis_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

# ============================================================
# FITNESS FUNCTION
# ============================================================

def fitness(params):
    n_estimators, contamination = int(params[0]), params[1]
    model = IsolationForest(
        n_estimators=n_estimators,
        contamination=contamination,
        random_state=42
    )
    preds = model.fit_predict(X_scaled)
    return np.mean(preds == 1)  # maximize legit stability

# ============================================================
# AIS INITIAL POPULATION
# ============================================================

population = [
    [random.randint(50, 300), random.uniform(0.01, 0.15)]
    for _ in range(12)
]

best_solution = None
best_score = -1

# ============================================================
# CSA LEVY FLIGHT
# ============================================================

def levy_flight():
    return np.random.normal(0, 1)

# ============================================================
# HYBRID AIS + CSA LOOP
# ============================================================

for _ in range(10):
    scores = [(p, fitness(p)) for p in population]
    scores.sort(key=lambda x: x[1], reverse=True)

    # AIS: Select top antibodies
    selected = [p for p, _ in scores[:5]]

    # Clone & mutate
    clones = []
    for s in selected:
        for _ in range(3):
            clone = [
                int(np.clip(s[0] + random.randint(-10, 10), 50, 300)),
                float(np.clip(s[1] + random.uniform(-0.02, 0.02), 0.01, 0.15))
            ]
            clones.append(clone)

    # CSA: Levy flight replacement
    for i in range(len(clones)):
        if random.random() < 0.25:
            clones[i][0] = int(np.clip(clones[i][0] + levy_flight() * 10, 50, 300))
            clones[i][1] = float(np.clip(clones[i][1] + levy_flight() * 0.01, 0.01, 0.15))

    population = clones

    for p in population:
        score = fitness(p)
        if score > best_score:
            best_score = score
            best_solution = p.copy()

# ============================================================
# FINAL HYBRID MODEL
# ============================================================

iso = IsolationForest(
    n_estimators=int(best_solution[0]),
    contamination=best_solution[1],
    random_state=42
)

pred = iso.fit_predict(X_scaled)
df["fraud_flag"] = np.where(pred == -1, "Fraud", "Legit")
y_binary = (pred == -1).astype(int)

with open(os.path.join(MODEL_DIR, "cis_fraud_model.pkl"), "wb") as f:
    pickle.dump(iso, f)

# ============================================================
# NEURAL VALIDATION MODEL
# ============================================================

nn = Sequential([
    Dense(32, activation="relu", input_shape=(X_scaled.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")
])

nn.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
history = nn.fit(X_scaled, y_binary, epochs=15, batch_size=8, verbose=0)

save_model(nn, os.path.join(MODEL_DIR, "cis_fraud_model.h5"))

# ============================================================
# GRAPHS (SHOW + SAVE)
# ============================================================

plt.plot(history.history["accuracy"])
plt.title("CIS Accuracy")
plt.savefig(os.path.join(GRAPH_DIR, "cis_accuracy.png"))
plt.show()

cm = confusion_matrix(y_binary, y_binary)
sns.heatmap(cm, annot=True, fmt="d")
plt.title("CIS Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "cis_heatmap.png"))
plt.show()

df["fraud_flag"].value_counts().plot(kind="bar")
plt.title("CIS Fraud vs Legit")
plt.savefig(os.path.join(GRAPH_DIR, "cis_comparison.png"))
plt.show()

plt.hist(df["degree"], bins=20)
plt.title("CIS Result Graph")
plt.savefig(os.path.join(GRAPH_DIR, "cis_result_graph.png"))
plt.show()

plt.scatter(df.index, y_binary)
plt.title("CIS Prediction Graph")
plt.savefig(os.path.join(GRAPH_DIR, "cis_prediction_graph.png"))
plt.show()

# ============================================================
# SAVE RESULTS
# ============================================================

df.to_csv(os.path.join(RESULT_DIR, "cis_results.csv"), index=False)

with open(os.path.join(RESULT_DIR, "cis_predictions.json"), "w") as f:
    json.dump(
        df[["project_id", "issuer_id", "fraud_flag"]].to_dict(orient="records"),
        f,
        indent=4
    )

# ============================================================
# METADATA
# ============================================================

meta = {
    "model": "Hybrid AIS + CSA (CIS)",
    "optimized_parameters": {
        "n_estimators": int(best_solution[0]),
        "contamination": best_solution[1]
    },
    "fraud_detected": int((df["fraud_flag"] == "Fraud").sum())
}

with open(os.path.join(META_DIR, "cis_metadata.yaml"), "w") as f:
    yaml.dump(meta, f)

print(" HYBRID AIS + CSA (CIS) PIPELINE COMPLETED SUCCESSFULLY")
```
