```python
import os
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import save_model

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Carbon Credit Fraud Detection"
DATA_PATH = os.path.join(BASE_DIR, "Table_5.5_Carbon_Stock_in_India_Forests.csv")

MODEL_DIR = os.path.join(BASE_DIR, "model")
RESULT_DIR = os.path.join(BASE_DIR, "results")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")
META_DIR = os.path.join(BASE_DIR, "metadata")

for d in [MODEL_DIR, RESULT_DIR, GRAPH_DIR, META_DIR]:
    os.makedirs(d, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.fillna(method="ffill", inplace=True)
df.reset_index(drop=True, inplace=True)

df["issuer_id"] = df.index % 10
df["project_id"] = df.index

# ============================================================
# GRAPH FEATURES
# ============================================================

G = nx.Graph()
for _, r in df.iterrows():
    G.add_edge(r["issuer_id"], r["project_id"])

df["degree"] = df["project_id"].apply(lambda x: G.degree(x))
df["issuer_degree"] = df["issuer_id"].apply(lambda x: G.degree(x))

# ============================================================
# FEATURE MATRIX
# ============================================================

num_cols = df.select_dtypes(include=["int64", "float64"]).columns
X = df[num_cols]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

with open(os.path.join(BASE_DIR, "his_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

# ============================================================
# FITNESS FUNCTION
# ============================================================

def fitness(params):
    n_estimators, contamination = int(params[0]), params[1]
    model = IsolationForest(
        n_estimators=n_estimators,
        contamination=contamination,
        random_state=42
    )
    preds = model.fit_predict(X_scaled)
    return np.mean(preds == 1)  # maximize legit consistency

# ============================================================
# AIS INITIAL POPULATION
# ============================================================

population = [
    [random.randint(50, 300), random.uniform(0.01, 0.15)]
    for _ in range(12)
]

best_solution = None
best_score = -1

# ============================================================
# HARMONY SEARCH PARAMETERS
# ============================================================

HMCR = 0.9   # Harmony Memory Considering Rate
PAR = 0.3    # Pitch Adjustment Rate
BW = 10      # Bandwidth

# ============================================================
# HYBRID AIS + HSA LOOP
# ============================================================

for _ in range(10):
    scores = [(p, fitness(p)) for p in population]
    scores.sort(key=lambda x: x[1], reverse=True)

    harmony_memory = [p for p, _ in scores[:5]]

    new_population = []

    for _ in range(15):
        if random.random() < HMCR:
            harmony = random.choice(harmony_memory).copy()

            if random.random() < PAR:
                harmony[0] = int(np.clip(harmony[0] + random.randint(-BW, BW), 50, 300))
                harmony[1] = float(np.clip(harmony[1] + random.uniform(-0.02, 0.02), 0.01, 0.15))
        else:
            harmony = [
                random.randint(50, 300),
                random.uniform(0.01, 0.15)
            ]

        new_population.append(harmony)

    population = new_population

    for p in population:
        score = fitness(p)
        if score > best_score:
            best_score = score
            best_solution = p.copy()

# ============================================================
# FINAL HYBRID MODEL
# ============================================================

iso = IsolationForest(
    n_estimators=int(best_solution[0]),
    contamination=best_solution[1],
    random_state=42
)

pred = iso.fit_predict(X_scaled)
df["fraud_flag"] = np.where(pred == -1, "Fraud", "Legit")
y_binary = (pred == -1).astype(int)

with open(os.path.join(MODEL_DIR, "his_fraud_model.pkl"), "wb") as f:
    pickle.dump(iso, f)

# ============================================================
# NEURAL VALIDATION MODEL
# ============================================================

nn = Sequential([
    Dense(32, activation="relu", input_shape=(X_scaled.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")
])

nn.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
history = nn.fit(X_scaled, y_binary, epochs=15, batch_size=8, verbose=0)

save_model(nn, os.path.join(MODEL_DIR, "his_fraud_model.h5"))

# ============================================================
# GRAPHS (SHOW + SAVE)
# ============================================================

plt.plot(history.history["accuracy"])
plt.title("HIS Accuracy")
plt.savefig(os.path.join(GRAPH_DIR, "his_accuracy.png"))
plt.show()

cm = confusion_matrix(y_binary, y_binary)
sns.heatmap(cm, annot=True, fmt="d")
plt.title("HIS Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "his_heatmap.png"))
plt.show()

df["fraud_flag"].value_counts().plot(kind="bar")
plt.title("HIS Fraud vs Legit")
plt.savefig(os.path.join(GRAPH_DIR, "his_comparison.png"))
plt.show()

plt.hist(df["degree"], bins=20)
plt.title("HIS Result Graph")
plt.savefig(os.path.join(GRAPH_DIR, "his_result_graph.png"))
plt.show()

plt.scatter(df.index, y_binary)
plt.title("HIS Prediction Graph")
plt.savefig(os.path.join(GRAPH_DIR, "his_prediction_graph.png"))
plt.show()

# ============================================================
# SAVE RESULTS
# ============================================================

df.to_csv(os.path.join(RESULT_DIR, "his_results.csv"), index=False)

with open(os.path.join(RESULT_DIR, "his_predictions.json"), "w") as f:
    json.dump(
        df[["project_id", "issuer_id", "fraud_flag"]].to_dict(orient="records"),
        f,
        indent=4
    )

# ============================================================
# METADATA
# ============================================================

meta = {
    "model": "Hybrid AIS + HSA (HIS)",
    "optimized_parameters": {
        "n_estimators": int(best_solution[0]),
        "contamination": best_solution[1]
    },
    "fraud_detected": int((df["fraud_flag"] == "Fraud").sum())
}

with open(os.path.join(META_DIR, "his_metadata.yaml"), "w") as f:
    yaml.dump(meta, f)

print(" HYBRID AIS + HSA (HIS) PIPELINE COMPLETED SUCCESSFULLY")
```
