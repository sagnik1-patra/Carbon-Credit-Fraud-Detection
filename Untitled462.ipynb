```python
import os
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import save_model

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Carbon Credit Fraud Detection"
DATA_PATH = os.path.join(BASE_DIR, "Table_5.5_Carbon_Stock_in_India_Forests.csv")

MODEL_DIR = os.path.join(BASE_DIR, "model")
RESULT_DIR = os.path.join(BASE_DIR, "results")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")
META_DIR = os.path.join(BASE_DIR, "metadata")

for d in [MODEL_DIR, RESULT_DIR, GRAPH_DIR, META_DIR]:
    os.makedirs(d, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.fillna(method="ffill", inplace=True)
df.reset_index(drop=True, inplace=True)

df["issuer_id"] = df.index % 10
df["project_id"] = df.index

# ============================================================
# GRAPH FEATURES
# ============================================================

G = nx.Graph()
for _, r in df.iterrows():
    G.add_edge(r["issuer_id"], r["project_id"])

df["degree"] = df["project_id"].apply(lambda x: G.degree(x))
df["issuer_degree"] = df["issuer_id"].apply(lambda x: G.degree(x))

# ============================================================
# FEATURE MATRIX
# ============================================================

num_cols = df.select_dtypes(include=["int64", "float64"]).columns
X = df[num_cols]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

with open(os.path.join(BASE_DIR, "psa_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

# ============================================================
# FITNESS FUNCTION
# ============================================================

def fitness(p):
    model = IsolationForest(
        n_estimators=int(p[0]),
        contamination=p[1],
        random_state=42
    )
    preds = model.fit_predict(X_scaled)
    return np.mean(preds == 1)

# ============================================================
# PSO INITIALIZATION
# ============================================================

n_particles = 12
particles = [[random.randint(50, 300), random.uniform(0.01, 0.15)] for _ in range(n_particles)]
velocities = [[0, 0] for _ in range(n_particles)]

pbest = particles.copy()
pbest_scores = [fitness(p) for p in particles]

gbest = pbest[np.argmax(pbest_scores)]
gbest_score = max(pbest_scores)

# PSO params
w, c1, c2 = 0.6, 1.5, 1.5

# ============================================================
# CSA LEVY FLIGHT
# ============================================================

def levy():
    return np.random.normal(0, 1)

# ============================================================
# HYBRID PSO + CSA LOOP
# ============================================================

for _ in range(10):
    for i in range(n_particles):
        for d in range(2):
            velocities[i][d] = (
                w * velocities[i][d]
                + c1 * random.random() * (pbest[i][d] - particles[i][d])
                + c2 * random.random() * (gbest[d] - particles[i][d])
            )
            particles[i][d] += velocities[i][d]

        particles[i][0] = int(np.clip(particles[i][0], 50, 300))
        particles[i][1] = float(np.clip(particles[i][1], 0.01, 0.15))

        score = fitness(particles[i])
        if score > pbest_scores[i]:
            pbest[i] = particles[i].copy()
            pbest_scores[i] = score

    gbest = pbest[np.argmax(pbest_scores)]
    gbest_score = max(pbest_scores)

    # CSA replacement (exploration)
    for i in range(n_particles):
        if random.random() < 0.25:
            particles[i][0] = int(np.clip(particles[i][0] + levy() * 10, 50, 300))
            particles[i][1] = float(np.clip(particles[i][1] + levy() * 0.01, 0.01, 0.15))

# ============================================================
# FINAL HYBRID MODEL
# ============================================================

iso = IsolationForest(
    n_estimators=int(gbest[0]),
    contamination=gbest[1],
    random_state=42
)

pred = iso.fit_predict(X_scaled)
df["fraud_flag"] = np.where(pred == -1, "Fraud", "Legit")
y_binary = (pred == -1).astype(int)

with open(os.path.join(MODEL_DIR, "psa_fraud_model.pkl"), "wb") as f:
    pickle.dump(iso, f)

# ============================================================
# NEURAL VALIDATION MODEL
# ============================================================

nn = Sequential([
    Dense(32, activation="relu", input_shape=(X_scaled.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")
])

nn.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
history = nn.fit(X_scaled, y_binary, epochs=15, batch_size=8, verbose=0)

save_model(nn, os.path.join(MODEL_DIR, "psa_fraud_model.h5"))

# ============================================================
# GRAPHS (SHOW + SAVE)
# ============================================================

plt.plot(history.history["accuracy"])
plt.title("PSA Accuracy")
plt.savefig(os.path.join(GRAPH_DIR, "psa_accuracy.png"))
plt.show()

cm = confusion_matrix(y_binary, y_binary)
sns.heatmap(cm, annot=True, fmt="d")
plt.title("PSA Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "psa_heatmap.png"))
plt.show()

df["fraud_flag"].value_counts().plot(kind="bar")
plt.title("PSA Fraud vs Legit")
plt.savefig(os.path.join(GRAPH_DIR, "psa_comparison.png"))
plt.show()

plt.hist(df["degree"], bins=20)
plt.title("PSA Result Graph")
plt.savefig(os.path.join(GRAPH_DIR, "psa_result_graph.png"))
plt.show()

plt.scatter(df.index, y_binary)
plt.title("PSA Prediction Graph")
plt.savefig(os.path.join(GRAPH_DIR, "psa_prediction_graph.png"))
plt.show()

# ============================================================
# SAVE RESULTS
# ============================================================

df.to_csv(os.path.join(RESULT_DIR, "psa_results.csv"), index=False)

with open(os.path.join(RESULT_DIR, "psa_predictions.json"), "w") as f:
    json.dump(
        df[["project_id", "issuer_id", "fraud_flag"]].to_dict(orient="records"),
        f,
        indent=4
    )

# ============================================================
# METADATA
# ============================================================

meta = {
    "model": "Hybrid PSO + CSA (PSA)",
    "optimized_parameters": {
        "n_estimators": int(gbest[0]),
        "contamination": gbest[1]
    },
    "fraud_detected": int((df["fraud_flag"] == "Fraud").sum())
}

with open(os.path.join(META_DIR, "psa_metadata.yaml"), "w") as f:
    yaml.dump(meta, f)

print(" HYBRID PSO + CSA (PSA) PIPELINE COMPLETED SUCCESSFULLY")
```
