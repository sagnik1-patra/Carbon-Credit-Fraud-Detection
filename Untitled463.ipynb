```python
import os
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import save_model

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Carbon Credit Fraud Detection"
DATA_PATH = os.path.join(BASE_DIR, "Table_5.5_Carbon_Stock_in_India_Forests.csv")

MODEL_DIR = os.path.join(BASE_DIR, "model")
RESULT_DIR = os.path.join(BASE_DIR, "results")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")
META_DIR = os.path.join(BASE_DIR, "metadata")

for d in [MODEL_DIR, RESULT_DIR, GRAPH_DIR, META_DIR]:
    os.makedirs(d, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.fillna(method="ffill", inplace=True)
df.reset_index(drop=True, inplace=True)

df["issuer_id"] = df.index % 10
df["project_id"] = df.index

# ============================================================
# GRAPH FEATURES
# ============================================================

G = nx.Graph()
for _, r in df.iterrows():
    G.add_edge(r["issuer_id"], r["project_id"])

df["degree"] = df["project_id"].apply(lambda x: G.degree(x))
df["issuer_degree"] = df["issuer_id"].apply(lambda x: G.degree(x))

# ============================================================
# FEATURE MATRIX
# ============================================================

num_cols = df.select_dtypes(include=["int64", "float64"]).columns
X = df[num_cols]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

with open(os.path.join(BASE_DIR, "gso_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

# ============================================================
# FITNESS FUNCTION
# ============================================================

def fitness(ind):
    model = IsolationForest(
        n_estimators=int(ind[0]),
        contamination=ind[1],
        random_state=42
    )
    preds = model.fit_predict(X_scaled)
    return np.mean(preds == 1)

# ============================================================
# GA INITIAL POPULATION
# ============================================================

population = [
    [random.randint(50, 300), random.uniform(0.01, 0.15)]
    for _ in range(14)
]

# ============================================================
# PSO STRUCTURES
# ============================================================

velocities = [[0, 0] for _ in population]
pbest = population.copy()
pbest_scores = [fitness(p) for p in population]

gbest = pbest[np.argmax(pbest_scores)]
gbest_score = max(pbest_scores)

# GA + PSO parameters
ELITE = 5
MUT_RATE = 0.3
w, c1, c2 = 0.6, 1.5, 1.5

# ============================================================
# HYBRID GA + PSO LOOP
# ============================================================

for _ in range(10):
    scored = [(p, fitness(p)) for p in population]
    scored.sort(key=lambda x: x[1], reverse=True)

    elites = [p for p, _ in scored[:ELITE]]
    offspring = []

    # GA crossover + mutation
    while len(offspring) < len(population) - ELITE:
        p1, p2 = random.sample(elites, 2)
        child = [
            int((p1[0] + p2[0]) / 2),
            (p1[1] + p2[1]) / 2
        ]

        if random.random() < MUT_RATE:
            child[0] = int(np.clip(child[0] + random.randint(-20, 20), 50, 300))
            child[1] = float(np.clip(child[1] + random.uniform(-0.03, 0.03), 0.01, 0.15))

        offspring.append(child)

    population = elites + offspring

    # PSO refinement
    for i, particle in enumerate(population):
        for d in range(2):
            velocities[i][d] = (
                w * velocities[i][d]
                + c1 * random.random() * (pbest[i][d] - particle[d])
                + c2 * random.random() * (gbest[d] - particle[d])
            )
            particle[d] += velocities[i][d]

        particle[0] = int(np.clip(particle[0], 50, 300))
        particle[1] = float(np.clip(particle[1], 0.01, 0.15))

        score = fitness(particle)
        if score > pbest_scores[i]:
            pbest[i] = particle.copy()
            pbest_scores[i] = score

    gbest = pbest[np.argmax(pbest_scores)]
    gbest_score = max(pbest_scores)

# ============================================================
# FINAL HYBRID MODEL
# ============================================================

iso = IsolationForest(
    n_estimators=int(gbest[0]),
    contamination=gbest[1],
    random_state=42
)

pred = iso.fit_predict(X_scaled)
df["fraud_flag"] = np.where(pred == -1, "Fraud", "Legit")
y_binary = (pred == -1).astype(int)

with open(os.path.join(MODEL_DIR, "gso_fraud_model.pkl"), "wb") as f:
    pickle.dump(iso, f)

# ============================================================
# NEURAL VALIDATION MODEL
# ============================================================

nn = Sequential([
    Dense(32, activation="relu", input_shape=(X_scaled.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")
])

nn.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
history = nn.fit(X_scaled, y_binary, epochs=15, batch_size=8, verbose=0)

save_model(nn, os.path.join(MODEL_DIR, "gso_fraud_model.h5"))

# ============================================================
# GRAPHS (SHOW + SAVE)
# ============================================================

plt.plot(history.history["accuracy"])
plt.title("GSO Accuracy")
plt.savefig(os.path.join(GRAPH_DIR, "gso_accuracy.png"))
plt.show()

cm = confusion_matrix(y_binary, y_binary)
sns.heatmap(cm, annot=True, fmt="d")
plt.title("GSO Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "gso_heatmap.png"))
plt.show()

df["fraud_flag"].value_counts().plot(kind="bar")
plt.title("GSO Fraud vs Legit")
plt.savefig(os.path.join(GRAPH_DIR, "gso_comparison.png"))
plt.show()

plt.hist(df["degree"], bins=20)
plt.title("GSO Result Graph")
plt.savefig(os.path.join(GRAPH_DIR, "gso_result_graph.png"))
plt.show()

plt.scatter(df.index, y_binary)
plt.title("GSO Prediction Graph")
plt.savefig(os.path.join(GRAPH_DIR, "gso_prediction_graph.png"))
plt.show()

# ============================================================
# SAVE RESULTS
# ============================================================

df.to_csv(os.path.join(RESULT_DIR, "gso_results.csv"), index=False)

with open(os.path.join(RESULT_DIR, "gso_predictions.json"), "w") as f:
    json.dump(
        df[["project_id", "issuer_id", "fraud_flag"]].to_dict(orient="records"),
        f,
        indent=4
    )

# ============================================================
# METADATA
# ============================================================

meta = {
    "model": "Hybrid GA + PSO (GSO)",
    "optimized_parameters": {
        "n_estimators": int(gbest[0]),
        "contamination": gbest[1]
    },
    "fraud_detected": int((df["fraud_flag"] == "Fraud").sum())
}

with open(os.path.join(META_DIR, "gso_metadata.yaml"), "w") as f:
    yaml.dump(meta, f)

print(" HYBRID GA + PSO (GSO) PIPELINE COMPLETED SUCCESSFULLY")
```
